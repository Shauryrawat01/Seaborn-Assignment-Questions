```python
#Question 1
#Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables
#Line plot:  Line plots are suitable for displaying trends over time or any other ordered variable.
#Histogram: Histograms are employed to represent the distribution of a single numerical variable.
#Bar Plot: Bar plots are effective for comparing categorical variables.
#Heatmap: Heatmaps are used to visualize the relationship between two categorical variables or one categorical and one numerical variable. 
```

```python
#Qusetion 2
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.show()
```

```python
#Question 3
x= sns.load_dataset("titanic")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=x, x='pclass', y='age')
plt.title('Age by Pclass')
plt.subplot(1, 2, 2)
sns.boxplot(data=x, x='pclass', y='fare')
plt.title('Fare by Pclass')
plt.tight_layout()
plt.show()
```

```python
#Question 4
y= sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6))
sns.histplot(data=y, x="price", hue="cut", bins=30, kde=True, palette="husl")
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

```python
#Question 5
iris= sns.load_dataset("iris")
sns.pairplot(iris, hue="species", palette="husl")
plt.title('Pair Plot of Iris Dataset with Species')
plt.show()
```
